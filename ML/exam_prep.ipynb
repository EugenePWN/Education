## First task

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('../ML/datasets/pulsar_stars_new.csv')
data
```

```python
data = data[(10 <= data['MIP']) & (data['MIP'] <= 100)]
```

```python
print(data['MIP'].mean())
print(data['MIP'].min())
```

```python
data = data.sort_values(by=['SIP'])
```

```python
X = data.drop(columns='TG')
y = data['TG']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33, stratify=y)
```

```python
X_train['STDC'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train, y_train)
X_train_scaler = scaler.transform(X_train)
```

```python
X_test_scaler = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train_scaler.T[1].mean()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, confusion_matrix

lgr = LogisticRegression()
lgr.fit(X_train_scaler, y_train)
pred = lgr.predict(X_test_scaler)
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
f1_score(pred, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaler, y_train)
pred = knn.predict(X_test_scaler)
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
f1_score(pred, y_test)
```

## Second task

```python
df_train = pd.read_csv('../ML/datasets/MNIST_train.csv')
df_test = pd.read_csv('../ML/datasets/MNIST_test.csv')
```

```python
df_train = df_train[(df_train['label'] == 1) | (df_train['label'] == 0)]
df_test = df_test[(df_test['label'] == 1) | (df_test['label'] == 0)]
```

```python
df_test
```

```python
X_train = df_train.drop(columns='label')
y_train = df_train['label']

X_test = df_test.drop(columns='label')
y_test = df_test['label']
```

```python
from sklearn.decomposition import PCA

pca = PCA(svd_solver='full')
pca.fit(X_train)
```

```python
array = pca.explained_variance_ratio_
res = 0
for i in array:
    res += i
    if res >= 0.9:
        idx = np.where(array == i)
        print(res, idx[0][0]+1)
        break
```

OR

```python
np.where(np.cumsum(array) >= 0.9)[0][0] + 1
```

```python
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
X_train[0][0]
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier(n_estimators=500, learning_rate=0.8, random_state=23, max_depth=2)
gbc.fit(X_train, y_train)
pred = gbc.predict(X_test)
```

```python
from sklearn.metrics import classification_report

tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
print(classification_report(pred, y_test, digits=4))
```
